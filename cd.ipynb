```python
import numpy as np
import matplotlib.pyplot as pp
import scipy.interpolate

class SimpleCDModel(object):
    """Generative model for the circilar dichroism spectrum"""
    num_x = 50
    
    def __init__(self):
        # these are n_points x 2 arrays of the (wavelength, theta)
        # data from figure 1 of Nature Protocols 1, 2527 - 2535 (2007)
        # which I'm using as "representitive CD specta" for each of three
        # different secondary structures
        alpha_data = np.loadtxt('alpha_helix_cd.dat')
        beta_data = np.loadtxt('antiparallel_belta_cd.dat')
        extended_data = np.loadtxt('extended_cd.dat')
        
        # this is the x axis
        x = np.linspace(max(np.min(alpha_data[:,0]), np.min(beta_data[:, 0]), np.min(extended_data[:, 0])),
                              min(np.max(alpha_data[:,0]), np.max(beta_data[:, 0]), np.max(extended_data[:, 0])),
                              num=self.num_x)
        self.x = x
        
        # interpolations of the three basis signals
        self._alpha = scipy.interpolate.interp1d(alpha_data[:, 0],  alpha_data[:, 1])(x)
        self._beta = scipy.interpolate.interp1d(beta_data[:, 0],  beta_data[:, 1])(x)
        self._extended = scipy.interpolate.interp1d(extended_data[:, 0],  extended_data[:, 1])(x)
        
        # plot stuff?
        pp.plot(x, self._alpha, label='alpha')
        pp.plot(x, self._beta, label='beta')
        pp.plot(x, self._extended, label='extended')
        pp.legend()

    def __call__(self, x):
        """
        Compute a simulated CD spectra for a protein represnted by
        three coordinates, the fraction alpha, beta and extended,
        which should be
           
        Parameters
        ----------
        x : array_like, length=3
            The fraction of the structure in each of alpha, beta, and extended.
        """
        if not len(x) == 3:
            raise ValueError('x should be length 3')
        raw_signal = self._alpha * x[0] + self._beta * x[1] + self._extended * x[2]
        noise = np.random.randn(len(raw_signal))
        
        return raw_signal + noise
            
        
simple_cd_model = SimpleCDModel()
#pp.plot(simple_cd([1,0,0]))
```

```python
# lets do some learning
from sklearn import svm

# build row-normalized matrix X of shape=(250, 3)
# this will simulate our observed proteins
n_train = 250
X = np.random.rand(n_train, 3)
X = (X.T / np.sum(X.T, axis=0)).T

# support vector regression predicts R^N -> R
# we want to predict the "cd spectrum", which is a
# function R -> R, so we're going to augment the feature space
# with a fourth input, which is the wavelength

X2 = np.zeros((n_train*SimpleCDModel.num_x, 4))
y2 = np.zeros(n_train*SimpleCDModel.num_x)
print X2.shape
print y2.shape

for i in range(n_train):
    spectra = simple_cd_model(X[i])
    for j in range(SimpleCDModel.num_x):
        X2[i*SimpleCDModel.num_x + j] = [X[i,0], X[i,1], X[i,2], j]
        y2[i*SimpleCDModel.num_x + j] = spectra[j]

print X2[[0,1]]
print y2[[0,1]]
clf = svm.SVR(kernel='rbf', epsilon=0.1)
clf.fit(X2, y2)
```

```python
# these spectra are all were generated only based on the 250 randomly sampled
# "proteins" in the dataset. they aren't actually aware that the signal is really
# generated based on a linear combination of the first three features.

x = np.arange(len(simple_cd_model.x))
y = clf.predict(np.array([np.ones_like(x), np.zeros_like(x), np.zeros_like(x), x]).T)
pp.plot(x,y, label='alpha')

y = clf.predict(np.array([np.zeros_like(x), np.ones_like(x), np.zeros_like(x), x]).T)
pp.plot(x, y, label='beta')

y = clf.predict(np.array([np.zeros_like(x), np.zeros_like(x), np.ones_like(x), x]).T)
pp.plot(x, y, label='extended')

y = clf.predict(np.array([0.5*np.ones_like(x), 0.5*np.ones_like(x), np.zeros_like(x), x]).T)
pp.plot(x, y, label='alpha + beta')

pp.legend()
```
